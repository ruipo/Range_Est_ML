```python
import keras

import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize

def encode(series):
    return pd.get_dummies(series.astype(str))
```

```python
cnnr_model = keras.models.load_model('cnnr_iter1.h5')
cnnr_model.summary()
```

```python
cnnc_model = keras.models.load_model('cnnc_iter3.h5')
cnnc_model.summary()
```

```python
#CNN Classification testing
#Import label data

temp_labels = np.loadtxt('vec_mat_clabels_icex_src_0.01train.csv',delimiter=",")
labels_t = []

for l in range(len(temp_labels)):
    labels_t.append(str(temp_labels[l]))

labels_t = np.array(labels_t)
labels_t = labels_t.ravel()

def encode(series):
    return pd.get_dummies(series.astype(str))

y_train = encode(labels_t)
labels = list(y_train.columns.values)
float_labels = [float(i) for i in labels_t]

#Import test data
features_test = np.loadtxt('vec_mat_features_icex_src_testh_norm.csv',delimiter=",")
temp_ytest = np.loadtxt('vec_mat_rlabels_icex_src_testh.csv',delimiter=",")
y_test= []

real_test = features_test[:,0::2]
imag_test = features_test[:,1::2]
X_test = np.zeros((features_test.shape[0],32,32,2))

for k in range(features_test.shape[0]):
    count = 0
    for i in range(32):
        for j in range(i,32):
            X_test[k,i,j,0] = real_test[k,count]
            X_test[k,i,j,1] = imag_test[k,count]
            
            if i!=j:
                X_test[k,j,i,0] = X_test[k,i,j,0]
                X_test[k,j,i,1] = -X_test[k,i,j,1]
            
            count = count + 1



temp_ytest = temp_ytest.ravel()
y_test = temp_ytest

predictions = cnnc_model.predict(X_test)

pred_labels = []
for i in np.argmax(predictions, axis=1):
    pred_labels.append(float(labels[i]))
```

```python
#sort output for plotting
inds = y_test.argsort()
sorted_y_test = y_test[inds]

floaty_pred = np.array(pred_labels).astype(np.float)
sorted_y_pred = floaty_pred[inds]

#classification plotting
sorted_y_predh = np.array([float(numeric_string) for numeric_string in sorted_y_pred])
sorted_y_testh = np.array([float(numeric_string) for numeric_string in sorted_y_test])
```

```python
#Plotting

fig = plt.figure(figsize=(20,10))
fig.suptitle('CNN-c Predictions', fontsize=20, fontweight='bold')
fig.text(0.09, 0.5, 'Predicted Distance (Km)', fontsize=20,va='center', rotation='vertical')

plt.subplot(2,3,4)
colors = sorted_y_predc-sorted_y_testc
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_testc,sorted_y_predc,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_testc,sorted_y_testc,'r',linewidth = 2)
plt.quiver(sorted_y_testc, sorted_y_testc, np.zeros(len(sorted_y_testc)), sorted_y_predc-sorted_y_testc,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('+0.1% to BL strength',fontsize=20)
plt.grid()

plt.subplot(2,3,5)
colors = sorted_y_predd-sorted_y_testd
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_testd,sorted_y_predd,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_testd,sorted_y_testd,'r',linewidth = 2)
plt.quiver(sorted_y_testd, sorted_y_testd, np.zeros(len(sorted_y_testd)), sorted_y_predd-sorted_y_testd,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('+0.25% to BL strength',fontsize=20)
plt.grid()

plt.subplot(2,3,6)
colors = sorted_y_prede-sorted_y_teste
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_teste,sorted_y_prede,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_teste,sorted_y_teste,'r',linewidth = 2)
plt.quiver(sorted_y_teste, sorted_y_teste, np.zeros(len(sorted_y_teste)), sorted_y_prede-sorted_y_teste,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('+0.5% to BL strength',fontsize=20)
plt.grid()

plt.subplot(2,3,3)
colors = sorted_y_predf-sorted_y_testf
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_testf,sorted_y_predf,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_testf,sorted_y_testf,'r',linewidth = 2)
plt.quiver(sorted_y_testf, sorted_y_testf, np.zeros(len(sorted_y_testf)), sorted_y_predf-sorted_y_testf,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('-0.1% to BL strength',fontsize=20)
plt.grid()

plt.subplot(2,3,2)
colors = sorted_y_predg-sorted_y_testg
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_testg,sorted_y_predg,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_testg,sorted_y_testg,'r',linewidth = 2)
plt.quiver(sorted_y_testg, sorted_y_testg, np.zeros(len(sorted_y_testg)), sorted_y_predg-sorted_y_testg,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('-0.25% to BL strength',fontsize=20)
plt.grid()

plt.subplot(2,3,1)
colors = sorted_y_predh-sorted_y_testh
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_testh,sorted_y_predh,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_testh,sorted_y_testh,'r',linewidth = 2)
plt.quiver(sorted_y_testh, sorted_y_testh, np.zeros(len(sorted_y_testh)), sorted_y_predh-sorted_y_testh,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('-0.5% to BL strength',fontsize=20)
plt.grid()

plt.savefig('/Users/Rui/Desktop/cnnc_test.eps')

plt.show()
```
