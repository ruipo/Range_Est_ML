```python
import keras

import numpy as np
import pandas as pd
import seaborn as sn
from sklearn import preprocessing
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize
import matplotlib.dates as mds

def encode(series):
    return pd.get_dummies(series.astype(str))
```

```python
cnnr_model = keras.models.load_model('cnnr_iter1.h5')
cnnr_model.summary()
cnnc_model = keras.models.load_model('cnnc_iter3.h5')
cnnc_model.summary()
pred_labels_allc = []
pred_labels_allr = []
```

```python
#CNN Classification testing
features_test = np.loadtxt('vec_mat_features_icex_src_icex16_hour8_norm.csv',delimiter=",")
temp_ytest = np.loadtxt('vec_mat_clabels_icex_src_test2_norm.csv',delimiter=",")
y_test= []

real_test = features_test[:,0::2]
imag_test = features_test[:,1::2]
X_test = np.zeros((features_test.shape[0],32,32,2))

for k in range(features_test.shape[0]):
    count = 0
    for i in range(32):
        for j in range(i,32):
            X_test[k,i,j,0] = real_test[k,count]
            X_test[k,i,j,1] = imag_test[k,count]
            
            if i!=j:
                X_test[k,j,i,0] = X_test[k,i,j,0]
                X_test[k,j,i,1] = -X_test[k,i,j,1]
            
            count = count + 1

for l in range(len(temp_ytest)):
    y_test.append(str(temp_ytest[l]))

y_test = np.array(y_test)
y_test = y_test.ravel()
label_test = y_test.ravel()

y_test = encode(y_test)
class_names = list(y_test.columns.values)


predictions = cnnc_model.predict(X_test)

pred_labels = []
for i in np.argmax(predictions, axis=1):
    pred_labels.append(float(class_names[i]))
    
pred_labels_allc = pred_labels_allc+pred_labels 
```

```python
#CNN Regression Testing
features_test = np.loadtxt('vec_mat_features_icex_src_icex16_hour8_norm.csv',delimiter=",")

real_test = features_test[:,0::2]
imag_test = features_test[:,1::2]
X_test = np.zeros((features_test.shape[0],32,32,2))

for k in range(features_test.shape[0]):
    count = 0
    for i in range(32):
        for j in range(i,32):
            X_test[k,i,j,0] = real_test[k,count]
            X_test[k,i,j,1] = imag_test[k,count]
            
            if i!=j:
                X_test[k,j,i,0] = X_test[k,i,j,0]
                X_test[k,j,i,1] = -X_test[k,i,j,1]
            
            count = count + 1


predictions = cnnr_model.predict(X_test)
pred_labels_allr = pred_labels_allr+predictions.tolist() 
```

```python
ep_startt = 1457852733 #38m
time = np.arange(0,len(pred_labels_allc)*10.92,10.92) + ep_startt
```

```python
time = np.arange(0,(len(pred_labels_allc))*10.92,10.92)
```

```python
window_size = 55

i = 0
moving_averagesc = []
while i < len(pred_labels_allc) - window_size + 1:
    this_window = pred_labels_allc[i : i + window_size]
    window_averagec = np.sum(this_window) / window_size
    moving_averagesc.append(window_averagec)
    i += 1

i = 0
moving_averagesr = []
while i < len(pred_labels_allr) - window_size + 1:
    this_window = pred_labels_allr[i : i + window_size]
    window_averager = np.sum(this_window) / window_size
    moving_averagesr.append(window_averager)
    i += 1
    
print(len(moving_averagesc))
print(len(moving_averagesr))
print(len(time))
print(len(pred_labels_allr))
print(len(pred_labels_allc))
```

```python
# Try Bokeh
from bokeh.io import output_file, show
from bokeh.layouts import column
from bokeh.plotting import figure
from bokeh.embed import components
from bokeh.models import DatetimeTickFormatter
from datetime import datetime
```

```python
output_file("icex38_test.html")

dates1 = [datetime.fromtimestamp(ts) for ts in mpf_time+14400]
dates2 = [datetime.fromtimestamp(ts) for ts in mpf_10minavg_time+14400]

p1 = figure(title="ICEX16 38m-depth MFP Predictions",plot_width=800, plot_height=250,x_axis_type="datetime",x_range=(dates1[0],dates1[-1]), y_range=(0,51))
p1.circle(dates1,mpf_pred,size=5, color="navy", alpha=0.5)
p1.line(dates2,mpf_10minavg_pred, color='red',line_width=2)
p1.xaxis.formatter=DatetimeTickFormatter(days="%m/%d %H:%M",
months="%m/%d %H:%M",
hours="%m/%d %H:%M",
minutes="%m/%d %H:%M")
p1.sizing_mode = 'scale_both'

dates1 = [datetime.fromtimestamp(ts) for ts in time+14400]
dates2 = [datetime.fromtimestamp(ts) for ts in time[int(window_size/2):int(-(window_size/2))]+14400]

p2 = figure(title="ICEX16 38m-depth CNN-c Predictions",plot_width=800, plot_height=250,x_axis_type="datetime",x_range=(dates1[0],dates1[-1]), y_range=(0,51))
p2.circle(dates1,pred_labels_allc,size=5, color="navy", alpha=0.5)
p2.line(dates2,moving_averagesc, color='red',line_width=2)
p2.yaxis.axis_label = "Predicted Distance (km)"
p2.xaxis.formatter=DatetimeTickFormatter(days="%m/%d %H:%M",
months="%m/%d %H:%M",
hours="%m/%d %H:%M",
minutes="%m/%d %H:%M")
p2.sizing_mode = 'scale_both'

p3 = figure(title="ICEX16 38m-depth CNN-r Predictions",plot_width=800, plot_height=250,x_axis_type="datetime",x_range=(dates1[0],dates1[-1]), y_range=(0,51))
p3.circle(dates1,pred_labels_allr,size=5, color="navy", alpha=0.5,legend_label='sample predictions')
p3.line(dates2,moving_averagesr, color='red',line_width=2,legend_label='10-min moving avg.')
p3.xaxis.axis_label = "Time (UTC)"
p3.legend.location = "bottom_right"
p3.xaxis.formatter=DatetimeTickFormatter(days="%m/%d %H:%M",
months="%m/%d %H:%M",
hours="%m/%d %H:%M",
minutes="%m/%d %H:%M")
p3.sizing_mode = 'scale_both'

plot = column(p1, p2, p3)
plot.sizing_mode = 'scale_both'
show(plot)
```

```python
#HTML Embedding
script, div = components(plot)
print(script)
print(div)
```
