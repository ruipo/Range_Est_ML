```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize
```

```python
mfp_pred_orig = np.loadtxt('mfp_test2_pred.csv',delimiter=",")
mfp_act_orig = np.loadtxt('mfp_test2_act.csv',delimiter=",")
mfp_pred_c = np.loadtxt('mfp_testc_pred.csv',delimiter=",") #+0.1% BL
mfp_act_c = np.loadtxt('mfp_testc_act.csv',delimiter=",")
mfp_pred_d = np.loadtxt('mfp_testd_pred.csv',delimiter=",") #+0.25% BL
mfp_act_d = np.loadtxt('mfp_testd_act.csv',delimiter=",")
mfp_pred_e = np.loadtxt('mfp_teste_pred.csv',delimiter=",") #+0.5% BL
mfp_act_e = np.loadtxt('mfp_teste_act.csv',delimiter=",")
mfp_pred_f = np.loadtxt('mfp_testf_pred.csv',delimiter=",") #-0.1% BL
mfp_act_f = np.loadtxt('mfp_testf_act.csv',delimiter=",")
mfp_pred_g = np.loadtxt('mfp_testg_pred.csv',delimiter=",") #-0.25% BL
mfp_act_g = np.loadtxt('mfp_testg_act.csv',delimiter=",")
mfp_pred_h = np.loadtxt('mfp_testh_pred.csv',delimiter=",") #-0.5% BL
mfp_act_h = np.loadtxt('mfp_testh_act.csv',delimiter=",")
```

```python
#Plotting
fig = plt.figure(figsize=(20,10))
fig.suptitle('MFP Predictions', fontsize=20, fontweight='bold')
fig.text(0.09, 0.5, 'Predicted Distance (Km)', fontsize=20,va='center', rotation='vertical')

plt.subplot(2,3,4)
colors = mfp_pred_c-mfp_act_c
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(mfp_act_c,mfp_pred_c,marker='.',c='k',alpha=0.75)
plt.plot(mfp_act_c,mfp_act_c,'r',linewidth = 2)
plt.quiver(mfp_act_c, mfp_act_c, np.zeros(len(mfp_act_c)), mfp_pred_c-mfp_act_c,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylabel('Predicted Distance (Km)',fontsize=20)
#plt.xlabel('Actual Distance (Km)',fontsize=20)
plt.title('+0.1% to BL Strength',fontsize=20)
plt.grid()

plt.subplot(2,3,5)
colors = mfp_pred_d-mfp_act_d
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(mfp_act_d,mfp_pred_d,marker='.',c='k',alpha=0.75)
plt.plot(mfp_act_d,mfp_act_d,'r',linewidth = 2)
plt.quiver(mfp_act_d, mfp_act_d, np.zeros(len(mfp_act_d)), mfp_pred_d-mfp_act_d,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylabel('Predicted Distance (Km)',fontsize=20)
plt.xlabel('Actual Distance (Km)',fontsize=20)
plt.title('+0.25% to BL Strength',fontsize=20)
plt.grid()

plt.subplot(2,3,6)
colors = mfp_pred_e-mfp_act_e
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(mfp_act_e,mfp_pred_e,marker='.',c='k',alpha=0.75)
plt.plot(mfp_act_e,mfp_act_e,'r',linewidth = 2)
plt.quiver(mfp_act_e, mfp_act_e, np.zeros(len(mfp_act_e)), mfp_pred_e-mfp_act_e,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylabel('Predicted Distance (Km)',fontsize=25)
#plt.xlabel('Actual Distance (Km)',fontsize=20)
plt.title('+0.5% to BL Strength',fontsize=20)
plt.grid()

plt.subplot(2,3,3)
colors = mfp_pred_f-mfp_act_f
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(mfp_act_f,mfp_pred_f,marker='.',c='k',alpha=0.75)
plt.plot(mfp_act_f,mfp_act_f,'r',linewidth = 2)
plt.quiver(mfp_act_f, mfp_act_f, np.zeros(len(mfp_act_f)), mfp_pred_f-mfp_act_f,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylabel('Predicted Distance (Km)',fontsize=20)
#plt.xlabel('Actual Distance (Km)',fontsize=20)
plt.title('-0.1% to BL Strength',fontsize=20)
plt.grid()

plt.subplot(2,3,2)
colors = mfp_pred_g-mfp_act_g
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(mfp_act_g,mfp_pred_g,marker='.',c='k',alpha=0.75)
plt.plot(mfp_act_g,mfp_act_g,'r',linewidth = 2)
plt.quiver(mfp_act_g, mfp_act_g, np.zeros(len(mfp_act_g)), mfp_pred_g-mfp_act_g,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylabel('Predicted Distance (Km)',fontsize=20)
#plt.xlabel('Actual Distance (Km)',fontsize=20)
plt.title('-0.25% to BL Strength',fontsize=20)
plt.grid()

plt.subplot(2,3,1)
colors = mfp_pred_h-mfp_act_h
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(mfp_act_h,mfp_pred_h,marker='.',c='k',alpha=0.75)
plt.plot(mfp_act_h,mfp_act_h,'r',linewidth = 2)
plt.quiver(mfp_act_h, mfp_act_h, np.zeros(len(mfp_act_h)), mfp_pred_h-mfp_act_h,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([3, 50, 3, 50],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.ylabel('Predicted Distance (Km)',fontsize=20)
#plt.xlabel('Actual Distance (Km)',fontsize=20)
plt.title('-0.5% to BL Strength',fontsize=20)
plt.grid()

plt.savefig('/Users/Rui/Desktop/mfp_test.eps')

plt.show()
```
