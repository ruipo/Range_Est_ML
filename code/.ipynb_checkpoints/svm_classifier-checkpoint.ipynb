```python
import numpy as np
import scipy as sp
import time
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import svm
from sklearn import datasets
from sklearn import preprocessing
from sklearn import model_selection as ms
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.utils.multiclass import unique_labels
import csv
```

```python
#import data and labels
'''
features = np.loadtxt('vec_mat_features_icex_src_0.1cl.csv',delimiter=",")
temp_labels = np.loadtxt('vec_mat_labels_icex_src_0.1cl.csv',delimiter=",")
labels = []

features = preprocessing.scale(features)
#features = features[np.nonzero(features)]
#features = features.reshape(1220,1055)
#KL = -sp.sum(features * sp.log(features), axis=1)
#X = np.array([np.real(KL), np.imag(KL)]).T

for l in range(len(temp_labels)):
    labels.append(str(temp_labels[l]))
    
labels = np.array(labels)
labels = labels.ravel()
'''


features = np.loadtxt('vec_mat_features_icex_src_0.01train_pca999p.csv',delimiter=",")
temp_labels = np.loadtxt('vec_mat_clabels_icex_src_0.01train.csv',delimiter=",")

labels_t = []
for l in range(len(temp_labels)):
    labels_t.append(str(temp_labels[l]))
    
labels_t = np.array(labels_t)
labels_t = labels_t.ravel()

def encode(series): 
    return pd.get_dummies(series.astype(str))

y_train = encode(labels_t)
class_names = list(y_train.columns.values)

labels = []

X_train = preprocessing.scale(features)

for l in range(len(temp_labels)):
    labels.append(str(temp_labels[l]))
    
labels = np.array(labels)
y_train = labels.ravel()
```

```python
#split dataset into training and test datasets

#X_train, X_test, y_train, y_test = ms.train_test_split(features, labels, test_size=0.6, random_state=0)


X_test = np.loadtxt('vec_mat_features_icex_src_test2_pca999p.csv',delimiter=",")
temp_ytest = np.loadtxt('vec_mat_clabels_icex_src_test2.csv',delimiter=",")
y_test= []

X_test = preprocessing.scale(X_test)

for l in range(len(temp_ytest)):
    y_test.append(str(temp_ytest[l]))
    
y_test = np.array(y_test)
y_test = y_test.ravel()


print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#cross validation

#rbf
C = [0.01, 0.1, 1, 10, 100]
gamma = [0.0001, 0.001, 0.01, 0.1]
scores_rbf = np.zeros((len(C),len(gamma)))

for i in range(len(C)):
    print(i)
    for j in range(len(gamma)):
        clf = svm.SVC(kernel='rbf',C = C[i], gamma = gamma[j], decision_function_shape='ovo', probability = False)
        scores = ms.cross_val_score(clf, X_train, y_train, cv=5)
        scores_rbf[i,j] = scores.mean()
        print("For C = %0.3f gamma = %0.4f , accuracy: %0.2f (+/- %0.2f)" % (C[i], gamma[j], scores.mean(), scores.std() * 2))

clf = svm.SVC(kernel='rbf',C = 10, gamma = 'scale', decision_function_shape='ovo', probability = False)
scores_rbf_scale = ms.cross_val_score(clf, X_train, y_train, cv=5)
print("For C = 10 gamma = scale , accuracy: %0.2f (+/- %0.2f)" % (scores_rbf_scale.mean(), scores_rbf_scale.std() * 2))
```

```python
#poly
C = [0.01, 0.1, 1, 10, 100]
degree = [3,5,7,9]
scores_poly = np.zeros((len(C),len(degree)))

for i in range(len(C)):
    print(i)
    for j in range(len(degree)):
        clf = svm.SVC(kernel='poly',C = C[i], degree = degree[j], gamma = 'scale', decision_function_shape='ovo', probability = False)
        scores = ms.cross_val_score(clf, X_train, y_train, cv=5)
        scores_poly[i,j] = scores.mean()
        print("For C = %0.3f degree = %0.4f , accuracy: %0.2f (+/- %0.2f)" % (C[i], degree[j], scores.mean(), scores.std() * 2))
```

```python
#Linear

clf = svm.SVC(kernel='linear',C = 1, decision_function_shape='ovo', probability = False)
scores_lin = ms.cross_val_score(clf, X_train, y_train, cv=5)
print("For C = %0.3f degree = %0.4f , accuracy: %0.2f (+/- %0.2f)" % (1, 1, scores_lin.mean(), scores_lin.std() * 2))
```

```python
#Sigmoid
C = [0.01, 0.1, 1, 10, 100]
gamma = [0.0001, 0.001, 0.01, 0.1]
scores_sig = np.zeros((len(C),len(gamma)))

for i in range(len(C)):
    print(i)
    for j in range(len(gamma)):
        clf = svm.SVC(kernel='sigmoid',C = C[i], gamma = gamma[j], decision_function_shape='ovo', probability = False)
        scores = ms.cross_val_score(clf, X_train, y_train, cv=5)
        scores_sig[i,j] = scores.mean()
        print("For C = %0.3f gamma = %0.4f , accuracy: %0.2f (+/- %0.2f)" % (C[i], gamma[j], scores.mean(), scores.std() * 2))

clf = svm.SVC(kernel='sigmoid',C = 10, gamma = 'scale', decision_function_shape='ovo', probability = False)
scores_sig_scale = ms.cross_val_score(clf, X_train, y_train, cv=5)
print("For C = 10 gamma = scale , accuracy: %0.2f (+/- %0.2f)" % (scores_sig_scale.mean(), scores_sig_scale.std() * 2))
```

```python
import matplotlib.pyplot as plt

def make_meshgrid(x, y, h=.02):
    """Create a mesh of points to plot in

    Parameters
    ----------
    x: data to base x-axis meshgrid on
    y: data to base y-axis meshgrid on
    h: stepsize for meshgrid, optional

    Returns
    -------
    xx, yy : ndarray
    """
    x_min, x_max = x.min() - 1, x.max() + 1
    y_min, y_max = y.min() - 1, y.max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy


def plot_contours(ax, clf, xx, yy, **params):
    """Plot the decision boundaries for a classifier.

    Parameters
    ----------
    ax: matplotlib axes object
    clf: a classifier
    xx: meshgrid ndarray
    yy: meshgrid ndarray
    params: dictionary of params to pass to contourf, optional
    """
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out



y = labels

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 10.0  # SVM regularization parameter
models = (svm.SVC(kernel='rbf', gamma='scale', C=C),
          svm.SVC(kernel='poly', degree=3, C=C))
models = (clf.fit(X, y) for clf in models)

# title for the plots
titles = ('SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel')

# Set-up 2x2 grid for plotting.
fig, sub = plt.subplots(1, 2)
plt.subplots_adjust(wspace=0.4, hspace=0.4)

X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)

for clf, title, ax in zip(models, titles, sub.flatten()):
    plot_contours(ax, clf, xx, yy,
                  cmap=plt.cm.coolwarm, alpha=0.8)
    ax.scatter(X0[0:20], X1[0:20],c='blue', cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.scatter(X0[20:40], X1[20:40],c='black', cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.scatter(X0[40:60], X1[40:60],c='red', cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.scatter(X0[60:80], X1[60:80],c='green', cmap=plt.cm.coolwarm, s=20, edgecolors='k')
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xlabel('Real KL')
    ax.set_ylabel('Image KL')
    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_title(title)

plt.show()
```
