```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.colors import Normalize
```

```python
sorted_y_pred1 = np.loadtxt('mfp_test1_pred.csv',delimiter=",") #216.5m
sorted_y_pred6 = np.loadtxt('mfp_test6_pred.csv',delimiter=",") #213.5m
sorted_y_pred7 = np.loadtxt('mfp_test7_pred.csv',delimiter=",") #219.5m
sorted_y_pred8 = np.loadtxt('mfp_test8_pred.csv',delimiter=",") #215.5m
sorted_y_pred9 = np.loadtxt('mfp_test9_pred.csv',delimiter=",") #217.5m

sorted_y_test1 = np.loadtxt('vec_mat_rlabels_swellex_109hz2_test.csv',delimiter=",")
sorted_y_test6 = np.loadtxt('vec_mat_rlabels_swellex_109hz2_test.csv',delimiter=",")
sorted_y_test7 = np.loadtxt('vec_mat_rlabels_swellex_109hz2_test.csv',delimiter=",")
sorted_y_test8 = np.loadtxt('vec_mat_rlabels_swellex_109hz2_test.csv',delimiter=",")
sorted_y_test9 = np.loadtxt('vec_mat_rlabels_swellex_109hz2_test.csv',delimiter=",")
```

```python
#Plotting

fig = plt.figure(figsize=(20,10))
fig.suptitle('MFP Predictions', fontsize=20, fontweight='bold')
fig.text(0.09, 0.5, 'Predicted Distance (Km)', fontsize=20,va='center', rotation='vertical')

plt.subplot(2,3,1)
colors = sorted_y_pred6-sorted_y_test6
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_test6,sorted_y_pred6,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_test6,sorted_y_test6,'r',linewidth = 2)
plt.quiver(sorted_y_test6, sorted_y_test6, np.zeros(len(sorted_y_test6)), sorted_y_pred6-sorted_y_test6,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([0, 10, 0, 10],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('Bottom Depth = 213.5m',fontsize=20)
plt.grid()

plt.subplot(2,3,2)
colors = sorted_y_pred8-sorted_y_test8
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_test8,sorted_y_pred8,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_test8,sorted_y_test8,'r',linewidth = 2)
plt.quiver(sorted_y_test8, sorted_y_test8, np.zeros(len(sorted_y_test8)), sorted_y_pred8-sorted_y_test8,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([0, 10, 0, 10],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('Bottom Depth = 215.5m',fontsize=20)
plt.grid()

plt.subplot(2,3,3)
colors = sorted_y_pred1-sorted_y_test1
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_test1,sorted_y_pred1,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_test1,sorted_y_test1,'r',linewidth = 2)
plt.quiver(sorted_y_test1, sorted_y_test1, np.zeros(len(sorted_y_test1)), sorted_y_pred1-sorted_y_test1,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([0, 10, 0, 10],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('Bottom Depth = 216.5m',fontsize=20)
plt.grid()

plt.subplot(2,3,4)
colors = sorted_y_pred9-sorted_y_test9
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_test9,sorted_y_pred9,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_test9,sorted_y_test9,'r',linewidth = 2)
plt.quiver(sorted_y_test9, sorted_y_test9, np.zeros(len(sorted_y_test9)), sorted_y_pred9-sorted_y_test9,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([0, 10, 0, 10],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
#plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('Bottom Depth = 217.5m',fontsize=20)
plt.grid()

plt.subplot(2,3,5)
colors = sorted_y_pred7-sorted_y_test7
norm = Normalize()
norm.autoscale(colors)
colormap = cm.rainbow
plt.scatter(sorted_y_test7,sorted_y_pred7,marker='.',c='k',alpha=0.75)
plt.plot(sorted_y_test7,sorted_y_test7,'r',linewidth = 2)
plt.quiver(sorted_y_test7, sorted_y_test7, np.zeros(len(sorted_y_test7)), sorted_y_pred7-sorted_y_test7,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
plt.axis([0, 10, 0, 10],fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel('Actual Distance (km)',fontsize=20)
plt.title('Bottom Depth = 219.5m',fontsize=20)
plt.grid()

# plt.subplot(2,3,1)
# colors = sorted_y_predh-sorted_y_testh
# norm = Normalize()
# norm.autoscale(colors)
# colormap = cm.rainbow
# plt.scatter(sorted_y_testh,sorted_y_predh,marker='.',c='k',alpha=0.75)
# plt.plot(sorted_y_testh,sorted_y_testh,'r',linewidth = 2)
# plt.quiver(sorted_y_testh, sorted_y_testh, np.zeros(len(sorted_y_testh)), sorted_y_predh-sorted_y_testh,color=colormap(norm(colors)),angles='xy', scale_units='xy', scale=1.,width=0.002)
# plt.axis([3, 50, 3, 50],fontsize=20)
# plt.xticks(fontsize=15)
# plt.yticks(fontsize=15)
# #plt.xlabel('Actual Distance (km)',fontsize=20)
# plt.title('-0.5% to BL strength',fontsize=20)
# plt.grid()

plt.savefig('/Users/Rui/Desktop/mfp_bdepth_test.eps')

plt.show()
```
